# Hashing

- mainly used to implement _**dictionaries and sets**_
  - **dictionaries**: key-value pairs
  - **sets**: collection of unique elements (only keys)
- **advantages**:
  - **Search**: $O(1)$ on average
  - **Insert**: $O(1)$ on average
  - **Delete**: $O(1)$ on average
  - these 3 operations or subset of them are required in many problems
- **comparison with other data structures**:
  - **arrays**:
    - if sorted, $O(\log n)$, $O(n)$ and $O(n)$ for search, insert and delete respectively
    - if unsorted, $O(n)$, $O(1)$ and $O(1)$ for search, insert and delete respectively
  - **Binary Search Tree**: self balancing BSTs like AVL, Red-Black, etc.
    - $O(\log n)$, $O(\log n)$ and $O(\log n)$ for search, insert and delete
  
> **Note**: Hashing is not suitable for problems where we need to maintain order of elements or range queries. For such problems, we can use arrays, BSTs, etc.

> **Note**: Hashing is not suitable for problems where we need to find closest elements, kth smallest/largest element, etc. For such problems, we can use heaps, BSTs, etc.

- not useful for
  - finding closest value (use AVL or Red-Black trees)
  - sorted data (use AVL or Red-Black trees)
  - prefix searching (use Trie)

## Applications of Hashing

- **Dictionaries**: store key-value pairs
- **Database Indexing**: to search records in database (either B/B++ trees or hashing)
- **Cryptography**: hash functions are used in cryptography
- **Caches**: to store data that is expensive to compute
- **Symbol Tables in Compilers/Interpreters**: to store variables, functions, etc.
- **Routers**: to store IP addresses
- **Getting data from databases**: to get data from databases quickly
- many more ...
